```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
fig.set_size_inches(18, 14)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([6.0742204, 11.2067787, 8.4654259, 6.467406, 15.9229475, 15.2994608])
plt.plot(x, y)
plt.xlabel("Option")
plt.ylabel("Time")
for i in range(len(x)):
    plt.annotate((x[i], y[i]), (x[i], y[i]))
```

```python
fig = plt.figure()
fig.set_size_inches(18, 14)
s = y[0] / y
plt.plot(x, s)
plt.xlabel("Number of nodes")
plt.ylabel("Speed up")
for i in range(len(x)):
    plt.annotate((x[i], s[i]), (x[i], s[i]))
```

```python
fig = plt.figure()
fig.set_size_inches(18, 14)
e = s / x
plt.plot(x, e)
plt.xlabel("Number of nodes")
plt.ylabel("Efficency")
for i in range(len(x)):
    plt.annotate((x[i], e[i]), (x[i], e[i]))
```
