```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
fig.set_size_inches(18, 14)
x = np.array([1, 2, 4, 8, 16, 32, 64, 128, 256, 512])
y = np.array([3.908, 2.966, 2.272, 2.086, 2.08, 2.106, 2.014, 2.144, 2.326, 2.226])
plt.plot(x, y)
plt.xlabel("Number of nodes")
plt.ylabel("Time")
for i in range(len(x)):
    plt.annotate((x[i], y[i]), (x[i], y[i]))
```

```python
fig = plt.figure()
fig.set_size_inches(18, 14)
s = y[0] / y
plt.plot(x, s)
plt.xlabel("Number of nodes")
plt.ylabel("Speed up")
for i in range(len(x)):
    plt.annotate((x[i], s[i]), (x[i], s[i]))
```

```python
fig = plt.figure()
fig.set_size_inches(18, 14)
e = s / x
plt.plot(x, e)
plt.xlabel("Number of nodes")
plt.ylabel("Efficency")
for i in range(len(x)):
    plt.annotate((x[i], e[i]), (x[i], e[i]))
```
